Ваша фінальна структура (запишіть її в змінну `df_registered_by_date`) даних має містити дві колонки: дату і кількість зареєстрованих користувачів на цей день.

```python
df['date'] = df.datetime.dt.date
```

```python
df_registered_by_date = df.pivot_table(index='date', values='registered', aggfunc='sum')
df_registered_by_date.head()
```

5. Тепер побудуйте лінійний графік з matplotlib для даних в `df_registered_by_date`.

  Додайте наступну частину до графіку аби зробити підписи за віссю ОХ більш читаємими

  `plt.xticks(rotation=45, ha='right')`

  Після побудови графіку проаналізуйте його та напишіть, які тенденції щодо зміни кількості зареєстрованих користувачів в часі ви бачите і чому так може бути?

```python
plt.figure(figsize=(20, 4))

plt.plot(df_registered_by_date.index, df_registered_by_date.registered)
plt.xticks(rotation=45, ha='right')
plt.title('Dynamics of registered users by date')
plt.xlabel('Date')
plt.ylabel('Number of registered users');
```

In this graph, we can observe the following trends:

A significant increase in client inflow for the company was noted starting from spring 2012. This may be due to increased marketing expenses or a rise in the number of bicycles in the city. These factors could explain such substantial growth.

There is a clear seasonality. From early spring (March) to mid-autumn (October), the demand for bicycles is higher compared to the off-season. As the weather warms up, there is less rain and fewer puddles, which increases the demand for bicycles.

The off-season can be considered the period from November to early to mid-March. As a hypothesis for verification: during this period, there are fewer tourists who come for vacation, which means there is less traffic congestion.

The historical minimum of registrations occurred in March 2011 (likely at the beginning of the project's development), while the historical maximum was in October 2012.
